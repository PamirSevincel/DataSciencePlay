### ML-specific Data Preprocessing

```python
# Convert categorical data into ordinal type. Only those that are more than binary 
# and/or have ordered categories are converted with this method. Otherwise, one-hot
# encoding is used.

from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

df_age = df[["AGE"]]
df_gender = df[["GENDER"]]
df_race = df[["RACE"]]
df_drivXP = df[["DRIVING_EXPERIENCE"]]
df_ed = df[["EDUCATION"]]
df_income = df[["INCOME"]]
df_vehYr = df[["VEHICLE_YEAR"]]
df_vehTy = df[["VEHICLE_TYPE"]]

age_encoded = ordinal_encoder.fit_transform(df_age)
gender_encoded = ordinal_encoder.fit_transform(df_gender)
race_encoded = ordinal_encoder.fit_transform(df_race)
drivXP_encoded = ordinal_encoder.fit_transform(df_drivXP)
vehYr_encoded = ordinal_encoder.fit_transform(df_vehYr)
vehTy_encoded = ordinal_encoder.fit_transform(df_vehTy)

df["AGE"] = age_encoded
df["GENDER"] = gender_encoded
df["RACE"] = race_encoded
df["DRIVING_EXPERIENCE"] = drivXP_encoded
df["VEHICLE_YEAR"] = vehYr_encoded
df["VEHICLE_TYPE"] = vehTy_encoded
```

Logistic Regression (R)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

model = LogisticRegression(solver='lbfgs') # add this inside to increase 1.0 precision & recall:  class_weight={0:1, 1:weight_minority_class}
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("Accuracy {0:.2f}%".format(100*accuracy_score(predictions, y_test)))
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
import matplotlib.pyplot as plt

def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.xlabel('Threshold')
    plt.legend(['Precision', 'Recall'])
    plt.grid(True)

plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
# plt.xlim([-1, 1])
plt.show()
```

Deep Learning | Tensorflow Keras (R)

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(9, kernel_initializer="uniform", activation="relu", input_dim=21))
model.add(Dense(1, kernel_initializer="uniform", activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# Display model summary and show parameters
model.summary()
```

```python
# Train the NN classifier

batch_size = 10
epochs = 25

history = model.fit(X_train,
                    y_train,
                    batch_size = batch_size,
                    epochs = epochs,
                    verbose = 1,
                    validation_data = (X_test, y_test))

score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

```python
# Plotting the loss charts

import matplotlib.pyplot as plt

history_dict = history.history

loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values) + 1)

line1 = plt.plot(epochs, val_loss_values, label='Validation/Test Loss')
line2 = plt.plot(epochs, loss_values, label='Training Loss')
plt.setp(line1, linewidth=2.0, marker='+', markersize=10.0)
plt.setp(line2, linewidth=2.0, marker='4', markersize=10.0)

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(False)
plt.legend()
plt.show()
```

```python
# Plotting the accuracy charts

history_dict = history.history

acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']
epochs = range(1, len(loss_values) + 1)

line1 = plt.plot(epochs, val_acc_values, label='Validation/Test Accuracy')
line2 = plt.plot(epochs, acc_values, label='Training Accuracy')
plt.setp(line1, linewidth=2.0, marker='+', markersize=10.0)
plt.setp(line2, linewidth=2.0, marker='4', markersize=10.0)

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.grid(False)
plt.legend()
plt.show()
```

```python
predictions = model.predict(X_test)
predictions = (predictions > 0.5)

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

Deeper NN Model (R)

```python
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Dropout

model2 = Sequential()

# Hidden Layer 1
model2.add(Dense(100, activation="relu", input_dim=21, kernel_regularizer=l2(0.01)))
model2.add(Dropout(0.3, noise_shape=None, seed=None))

# Hidden Layer 2
model2.add(Dense(250, activation="relu", kernel_regularizer=l2(0.01)))
model2.add(Dropout(0.3, noise_shape=None, seed=None))

# Hidden Layer 3
model2.add(Dense(500, activation="relu", kernel_regularizer=l2(0.01)))
model2.add(Dropout(0.3, noise_shape=None, seed=None))

# Hidden Layer 4
model2.add(Dense(250, activation="relu", kernel_regularizer=l2(0.01)))
model2.add(Dropout(0.3, noise_shape=None, seed=None))

# Hidden Layer 5
model2.add(Dense(100, activation="relu", kernel_regularizer=l2(0.01)))
model2.add(Dropout(0.3, noise_shape=None, seed=None))

model2.add(Dense(1, activation='sigmoid'))

model2.summary()
```

```python
model2.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# Train the deeper NN classifier

batch_size = 10
epochs = 25

history = model2.fit(X_train,
                    y_train,
                    batch_size = batch_size,
                    epochs = epochs,
                    verbose = 1,
                    validation_data = (X_test, y_test))

score = model2.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

```python
# Plotting the loss charts

history_dict = history.history

loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values) + 1)

line1 = plt.plot(epochs, val_loss_values, label='Validation/Test Loss')
line2 = plt.plot(epochs, loss_values, label='Training Loss')
plt.setp(line1, linewidth=2.0, marker='+', markersize=10.0)
plt.setp(line2, linewidth=2.0, marker='4', markersize=10.0)

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(False)
plt.legend()
plt.show()
```

```python
# Plotting the accuracy charts

history_dict = history.history

acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']
epochs = range(1, len(loss_values) + 1)

line1 = plt.plot(epochs, val_acc_values, label='Validation/Test Accuracy')
line2 = plt.plot(epochs, acc_values, label='Training Accuracy')
plt.setp(line1, linewidth=2.0, marker='+', markersize=10.0)
plt.setp(line2, linewidth=2.0, marker='4', markersize=10.0)

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.grid(False)
plt.legend()
plt.show()
```
